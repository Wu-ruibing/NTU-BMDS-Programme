```python
import numpy as np
import math as m
from sympy import *
import matplotlib.pyplot as plt

xarray=[]
yarray=[]
zarray=[]
warray=[]
array=[]

def f1_val(x1,x2,x3,x4):
    a=-100*x1*x3+900*x2
    return a

def f2_val(x1,x2,x3,x4):
    a=100*x1*x3-900*x2
    return a

def f3_val(x1,x2,x3,x4):
    a=-100*x1*x3+600*x2
    return a

def f4_val(x1,x2,x3,x4):
    a=150*x2
    return a

```

```python
xarray=[]
yarray=[]
zarray=[]
warray=[]
parray=[]
array=[]
def RK4():
    h=0.0005
    a=0
    x1=1
    x2=0
    x3=10
    x4=0
    while a<=0.02:
        array.append(a)
        xarray.append(x1)
        yarray.append(x2)
        zarray.append(x3)
        warray.append(x4)
       
        
        a+=h

        f1=f1_val(x1,x2,x3,x4) #第一步
        m1=x1+f1*h/2
        g1=f2_val(x1,x2,x3,x4)
        n1=x2+g1*h/2
        p1=f3_val(x1,x2,x3,x4)
        q1=x3+p1*h/2
        r1=f4_val(x1,x2,x3,x4)
        s1=x4+r1*h/2
        
        if m1 <= 0:
            m1 = 0
        if n1 <= 0:
            n1 = 0
        if q1 <= 0:
            q1 = 0
        if s1 <= 0:
            s1 = 0
        
        
        f2=f1_val(m1,n1,q1,s1) #第二步
        m2=x1+f2*h/2
        g2=f2_val(m1,n1,q1,s1)
        n2=x2+g2*h/2
        p2=f3_val(m1,n1,q1,s1)
        q2=x3+p2*h/2
        r2=f4_val(m1,n1,q1,s1)
        s2=x4+r2*h/2
         
        if m2 <= 0:
            m2 = 0
        if n2 <= 0:
            n2 = 0
        if q2 <= 0:
            q2 = 0
        if s2 <= 0:
            s2 = 0
    
 
        f3=f1_val(m2,n2,q2,s2) #第三步
        m3=x1+f3*h
        g3=f2_val(m2,n2,q2,s2)
        n3=x2+g3*h
        p3=f3_val(m2,n2,q2,s2)
        q3=x3+p3*h
        r3=f4_val(m2,n2,q2,s2)
        s3=x4+r3*h
        
        if m3 <= 0:
            m3 = 0
        if n3 <= 0:
            n3 = 0
        if q3 <= 0:
            q3 = 0
        if s3 <= 0:
            s3 = 0
        
        f4=f1_val(m3,n3,q3,s3) #第四步
        g4=f2_val(m3,n3,q3,s3)
        p4=f3_val(m3,n3,q3,s3)
        r4=f4_val(m3,n3,q3,s3)

        if f4 <= 0:
            f4 = 0
        if g4 <= 0:
            g4 = 0
        if p4 <= 0:
            p4 = 0
        if r4 <= 0:
            r4 = 0
        
      
        
        
        x1=x1+(f1+2*f2+2*f3+f4)*h/6
        x2=x2+(g1+2*g2+2*g3+g4)*h/6
        x3=x3+(p1+2*p2+2*p3+p4)*h/6
        x4=x4+(r1+2*r2+2*r3+r4)*h/6
        pelement=150*x2
        parray.append(pelement)
    
    
def main():
    RK4()
    for i in xarray:
        print(i)
    print("-------------")
    for i in yarray:
        print(i)
    print("-------------")
    for i in zarray:
        print(i)
    print("-------------")
    for i in warray:
        print(i)

if __name__ == "__main__":
    main()
```
